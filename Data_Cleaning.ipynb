# Removing Puncuation

```python
#Need to remove '$' from production_budget, domestic_gross, and worldwide_gross values
#try changing to a string first 
#should work if production_budget was a string
```

```python
#making prod_budget a string
movie_df['production_budget'] = movie_df['production_budget'].astype(str)
```

```python
df_obj = movie_df.select_dtypes(['object'])
```

```python
#Get rid of all the '$'
movie_df[df_obj.columns] = df_obj.apply(lambda x: x.str.strip('$'))
```

```python
#get rid of all ','

movie_df['production_budget']=movie_df['production_budget'].str.replace(',','')
movie_df['worldwide_gross']=movie_df['worldwide_gross'].str.replace(',','')
movie_df['domestic_gross']=movie_df['domestic_gross'].str.replace(',','')
```

# Changing dtype of Number Strings

```python
#changing production_budget to float 
movie_df['production_budget'] = movie_df['production_budget'].astype(float)
movie_df['worldwide_gross'] = movie_df['worldwide_gross'].astype(float)
movie_df['domestic_gross'] = movie_df['domestic_gross'].astype(float)
```

# Seperation of release month day and year 

```python
#month date and year in seperate columns

movie_df[['release_month','release_day','release_year']] = movie_df.release_date.str.split(expand=True)
```

```python
#get rid of the ',' in release_day

movie_df['release_day']=movie_df['release_day'].str.replace(',','')
```
